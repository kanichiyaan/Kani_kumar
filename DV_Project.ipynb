```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
#display first 5 datas on flights_data.csv by using head()
 
fs = pd.read_csv('/content/flights_data.csv') # flights_data.csv as fs
print(fs.shape)
fs.head()
```

```python
# Display the datas in the form of Bar Chart

sb.countplot(data =fs,x='Source'); # add ; at end of code to hide the metadata
```

```python
#color palate value [0] is blue colour
bc = sb.color_palette()[0]  # color_palate as bc ( base color )
sb.countplot(data= fs,x='Source', color = bc);
```

```python
#color palate value [1] is orange colour
bc = sb.color_palette()[1]  # color_palate as bc ( base color )
sb.countplot(data= fs,x='Source', color = bc);
```

```python
#color palate value [2] is green colour
bc = sb.color_palette()[2]  # color_palate as bc ( base color )
sb.countplot(data= fs,x='Source', color = bc);
```

```python
#color palate value [3] is red colour
bc = sb.color_palette()[3]  # color_palate as bc ( base color )
sb.countplot(data= fs,x='Source', color = bc);
```

```python
#
bc = sb.color_palette()[4]  # color_palate as bc ( base color )
so = fs['Source'].value_counts().index # source order as so
print(so)
sb.countplot(data= fs,x='Source', color = bc, order = so);
```

```python
#Additional variations 
# data of Airlines is in x - axis
bc = sb.color_palette()[2]  # color_palate as bc ( base color )
sb.countplot(data= fs,x='Airline', color = bc)
plt.xticks(rotation=90); # use rotation to reduce overlapping of texts
```

```python
#Additional variations 
# data of Airlines is in y - axis
bc = sb.color_palette()[2]  # color_palate as bc ( base color )
sb.countplot(data= fs,y='Airline', color = bc);
```

```python
#Count missing data

fs.isna().sum()
```

```python
md = fs.isna().sum() # missing data as md
bc = sb.color_palette()[2]  # color_palate as bc ( base color )
sb.barplot(md.index.values, md, color = bc)
plt.xticks(rotation=90); # use rotation to reduce overlapping of texts
plt.ylabel('Number of missing data', fontsize = 12);
```

```python
#Pie chart

sc = fs['Destination'].value_counts() #sorted counts as sc
plt.pie(sc, labels = sc.index, startangle = 90, counterclock = False)
plt.axis('square')
plt.title('Flight Destination\'s');
```

```python

sc = fs['Destination'].value_counts() #sorted counts as sc
plt.pie(sc, labels = sc.index, startangle = 90, counterclock = False, wedgeprops = {'width' : 0.4})
plt.axis('square')
plt.title('Flight Destination\'s');
```

```python
#Histograms
bc = sb.color_palette()[2]
plt.hist(data = fs, x = 'Duration(minutes)' , color = bc);
```

```python
bc = sb.color_palette()[2]
plt.hist(data = fs , x = 'Price' , bins = 20, color = bc);
```

```python
bc = sb.color_palette()[2]
bin = np.arange(0, fs['Price'].max()+1, 1200)
plt.hist(data = fs, x ='Price', bins = bin , color = bc)
plt.show();
```

```python
sb.distplot(fs['Price'], color = bc);
```

```python
sb.distplot(fs['Price'], kde = False, color = bc);
```

```python
bins = np.arange(0, fs['Price'].max()+1, 1200)
sb.distplot(fs['Price'], bins = bins, kde = False, hist_kws={'alpha':0.8}, color = bc);
```
